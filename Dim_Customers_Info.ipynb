```
from pyspark.sql.readwriter import DataFrameWriter

customer_result = spark.sql("select c_id as Customer_Id,c_unique_id as Customer_Unique_Id,c_name as Customer_Name,customer_pincode as Customer_Pincode,count(or_id) as Num_Of_Orders \
from global_temp.customer_info \
left outer join global_temp.orders \
on c_unique_id = customer_unique_id \
group by or_id,c_id,c_unique_id,c_name,customer_pincode")

customer_result = customer_result.coalesce(1)

customer_result.write.format('csv').mode('overwrite').option("header","true").save('/mnt/salesanalytic/Sales/Transformed_Data/Customers_Info/sparksqlresult/')
```
