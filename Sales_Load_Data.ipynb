```
from pyspark.sql.types import StringType,StructField,StructType,IntegerType,LongType,FloatType,DoubleType,TimestampType
```

```
cust_sch = [StructField('c_id',IntegerType(),nullable=False),
            StructField('c_unique_id',IntegerType(),nullable=False),
            StructField('c_name',StringType(),nullable=False),
            StructField('Customer_Account_Created_Timestamp',TimestampType(),nullable=True),
            StructField('customer_pincode',DoubleType(),nullable=False),
            StructField('customer_city',StringType(),nullable=True),
            StructField('customer_mobile_number',StringType(),nullable=False)]

customers_schema = StructType(fields=cust_sch)

pr_schema = [StructField('product_id',IntegerType(),nullable=False),
             StructField('product_name',StringType(),nullable=False),
             StructField('product_unit_cost',DoubleType(),nullable=False),
             StructField('product_max_discount_price',DoubleType(),nullable=True),
             StructField('product_seller_id',StringType(),nullable=False),
             StructField('product_created_timestamp',TimestampType(),nullable=True)]

products_schema = StructType(fields=pr_schema)

sell_sch = [StructField('seller_id',StringType(),nullable=False),
            StructField('seller_organization_name',StringType(),nullable=False),
            StructField('seller_state',StringType(),nullable=False),
            StructField('seller_city',StringType(),nullable=False),
            StructField('seller_address',StringType(),nullable=True),StructField('sellers_created_timestamp',TimestampType(),nullable=True)]

seller_schema = StructType(fields=sell_sch)
```

```
df_customers = spark.read.csv(path='/mnt/salesanalytic/Incremental/Customers_ADLS_Stage1/customers.csv',inferSchema="false",header='true',schema=customers_schema)
df_products = spark.read.csv("/mnt/salesanalytic/Historical/products_ADLS_Stage1/products.csv", header="true",inferSchema="false",schema=products_schema)
sf_sellers = spark.read.csv("/mnt/salesanalytic/Historical/Sellers_ADLS_Stage1/sellers.csv",header="true",inferSchema="false",schema=seller_schema)
df_orders = spark.read.csv('/mnt/salesanalytic/Incremental/orders_ADLS_Stage1/orders.csv',inferSchema='false',header='true')
df_payments_lookup = spark.read.csv("/mnt/salesanalytic/Historical/payments_Lookup_ADLS_Stage1/payment_types_lookup.csv",header='true',inferSchema='true')
df_payments = spark.read.csv('/mnt/salesanalytic/Incremental/payments_ADLS_Stage1/payment_information.csv',header='true',inferSchema='true')
```

```
df_orders.createOrReplaceGlobalTempView('orders')
df_customers.createOrReplaceGlobalTempView('customer_info')
df_products.createOrReplaceGlobalTempView("products")
sf_sellers.createOrReplaceGlobalTempView("sellers")
df_payments.createOrReplaceGlobalTempView('payments')
df_payments_lookup.createOrReplaceGlobalTempView('payments_lookup')
```
